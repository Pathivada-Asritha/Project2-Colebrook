```python
import numpy as np
import matplotlib.pyplot as plt

def colebrook_approx(friction_factor, roughness_ratio, reynolds_num, iterations):
    if (iterations > 0):
        friction_factor = colebrook_approx(friction_factor, roughness_ratio, reynolds_num, iterations - 1)

    denominator = -2 * np.log10((roughness_ratio / 3.7) + (2.51 / (reynolds_num * np.sqrt(friction_factor))))
    return 1 / (denominator ** 2)

def laminar_flow(reynolds_num):
    return 64 / reynolds_num

reynolds_vals = np.arange(2300, 10**8, step=1000)
roughness_vals = np.array([5e-2, 4e-2, 3e-2, 2e-2, 1e-2, 8e-3, 6e-3, 4e-3, 2e-3, 1e-3, 8e-4, 6e-4, 4e-4, 2e-4, 1e-4, 5e-5, 1e-5])
reynolds_vals, roughness_vals = np.meshgrid(reynolds_vals, roughness_vals)
friction_vals = colebrook_approx(0.03, roughness_vals, reynolds_vals, 10)

reynolds_lam = np.linspace(640, 10**3.4, 2)
friction_lam = laminar_flow(reynolds_lam)

plt.figure(figsize=(10, 12))
for i in range(reynolds_vals.shape[0]):
    plt.plot(reynolds_vals[i, :], friction_vals[i, :], linewidth=0.7, color='b')

plt.text(0.8e5, 0.0913, "Turbulent Flow", fontsize=18, rotation=0)

plt.plot(reynolds_lam, friction_lam, linewidth=0.7, color='g')
plt.text(0.7e3, 0.0522, "Laminar Flow (64/Re)", fontsize=12, rotation=-85)

plt.xscale("log")
plt.grid(which="both", linestyle="--", linewidth=0.7, color='k')
plt.xlabel("Reynolds number, Re")
plt.ylabel("Friction factor, f")
plt.title("Colebrook Equation Approximation")
ax = plt.gca()
ax.set_yticks(np.linspace(0.0, 0.1, 24))
ax2 = ax.twinx()
ax2.set_ylim(ax.get_ylim())  # Match limits with primary y-axis
ax2.set_yticks(friction_vals[:, -1])  # Place ticks at friction factor values
ax2.set_yticklabels([f"{val:.0e}" for val in roughness_vals[:, 1]])  # Format labels
ax2.set_ylabel("Relative roughness, e/D")

plt.show()
```
